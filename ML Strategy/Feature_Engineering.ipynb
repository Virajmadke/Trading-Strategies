```python
def Add_Indicators(Data):
    EMA8  = EMAIndicator(close = Data["Close"], window = 8, fillna = True)
    EMA13  = EMAIndicator(close = Data["Close"], window = 13, fillna = True)
    EMA21  = EMAIndicator(close = Data["Close"], window = 21, fillna = True)
    EMA34 = EMAIndicator(close = Data["Close"], window = 34, fillna = True)
    EMA55 = EMAIndicator(close = Data["Close"], window = 55, fillna = True)
    EMA89  = EMAIndicator(close = Data["Close"], window = 89, fillna = True)
    EMA144  = EMAIndicator(close = Data["Close"], window = 144, fillna = True)
    EMA233  = EMAIndicator(close = Data["Close"], window = 233, fillna = True)


    Data["EMA8"]  = EMA8.ema_indicator()
    Data["EMA13"]  = EMA13.ema_indicator()
    Data["EMA21"]  = EMA21.ema_indicator()
    Data["EMA34"] = EMA34.ema_indicator()
    Data["EMA55"] = EMA55.ema_indicator()
    Data["EMA89"]  = EMA89.ema_indicator()
    Data["EMA144"]  = EMA144.ema_indicator()
    Data["EMA233"]  = EMA233.ema_indicator()

    return Data
```

```python
%%time
Features = pd.DataFrame()
Dates= np.array([])
FivePercent= np.array([])
EightPercent= np.array([])
ThirteenPercent= np.array([])
AdvancedRatio= np.array([])
Declining= np.array([])
frac_EMA8 = np.array([])
frac_EMA13 = np.array([])
frac_EMA21 = np.array([])
frac_EMA34 = np.array([])
frac_EMA55 = np.array([])
frac_EMA89 = np.array([])
frac_EMA144 = np.array([])
frac_EMA233 = np.array([])

for i in range(1, Data.shape[0]):
    winners = 1
    losers = 1
    Per_5 = 1
    Per_8 = 1
    Per_13 = 1
    AD_Ratio = 1

    f_EMA8 = 0
    f_EMA13 = 0
    f_EMA21 = 0
    f_EMA34 = 0
    f_EMA55 = 0
    f_EMA89 = 0
    f_EMA144 = 0
    f_EMA233 = 0
    for ticker in Tickers:
        if(math.isnan(Data[ticker]["Close"][i]) != True):

            temp = Data[ticker]
            temp = Add_Indicators(temp)
            temp = temp[temp.index == Data[ticker].index[i]]

            if(temp["Close"][0] > temp["EMA8"][0]):
                f_EMA8 = f_EMA8 +  1

            if(temp["Close"][0] > temp["EMA13"][0]):
                f_EMA13 = f_EMA13 +  1

            if(temp["Close"][0] > temp["EMA21"][0]):
                f_EMA21 = f_EMA21 +  1

            if(temp["Close"][0] > temp["EMA34"][0]):
                f_EMA34 = f_EMA34 +  1

            if(temp["Close"][0] > temp["EMA55"][0]):
                f_EMA55 = f_EMA55 +  1

            if(temp["Close"][0] > temp["EMA89"][0]):
                f_EMA89 = f_EMA89 +  1

            if(temp["Close"][0] > temp["EMA144"][0]):
                f_EMA144 = f_EMA144 +  1

            if(temp["Close"][0] > temp["EMA233"][0]):
                f_EMA233 = f_EMA233 +  1


            if(Data[ticker]["Close"][i] >= Data[ticker]["Close"][i-1]):
                winners = winners + 1
                if((Data[ticker]["Close"][i] - Data[ticker]["Close"][i-1])/Data[ticker]["Close"][i-1] >= 0.05):
                    if((Data[ticker]["Close"][i] - Data[ticker]["Close"][i-1])/Data[ticker]["Close"][i-1] >= 0.08):
                        if((Data[ticker]["Close"][i] - Data[ticker]["Close"][i-1])/Data[ticker]["Close"][i-1] >= 0.13):
                            Per_13 = Per_13 + 1
                            continue
                        Per_8 = Per_8 + 1
                        continue        
                    Per_5 = Per_5 + 1
                    continue       
                           
            if(Data[ticker]["Close"][i] < Data[ticker]["Close"][i-1]):
                losers = losers + 1
                
    AD_Ratio = winners/losers
    Per_5  = Per_5/(winners + losers)
    Per_8  = Per_8/(winners + losers) 
    Per_13 = Per_13/(winners + losers)

    f_EMA8   = f_EMA8/(winners + losers)
    f_EMA13  = f_EMA13/(winners + losers)
    f_EMA21  = f_EMA21/(winners + losers)
    f_EMA34  = f_EMA34/(winners + losers)
    f_EMA55  = f_EMA55/(winners + losers)
    f_EMA89  = f_EMA89/(winners + losers)
    f_EMA144 = f_EMA144/(winners + losers)
    f_EMA233 = f_EMA233/(winners + losers)

    Dates= np.append(Dates, Data.index[i])
    FivePercent= np.append(FivePercent, Per_5)
    EightPercent= np.append(EightPercent, Per_8)
    ThirteenPercent= np.append(ThirteenPercent, Per_13)
    AdvancedRatio= np.append(AdvancedRatio, AD_Ratio)
    Declining= np.append(Declining, losers)

    frac_EMA8 = np.append(frac_EMA8, f_EMA8)
    frac_EMA13 = np.append(frac_EMA13, f_EMA13)
    frac_EMA21 = np.append(frac_EMA21, f_EMA21)
    frac_EMA34 = np.append(frac_EMA34, f_EMA34)
    frac_EMA55 = np.append(frac_EMA55, f_EMA55)
    frac_EMA89 = np.append(frac_EMA89, f_EMA89)
    frac_EMA144 = np.append(frac_EMA144, f_EMA144)
    frac_EMA233 = np.append(frac_EMA233, f_EMA233)
    


Features["Date"] = Dates
Features["AD_Ratio"] = AdvancedRatio
Features["5%_Up"] = FivePercent
Features["8%_Up"] = EightPercent
Features["13%_Up"] = ThirteenPercent
Features["Down"] = Declining
Features["Frac_EMA8"]   = frac_EMA8
Features["Frac_EMA13"]  = frac_EMA13
Features["Frac_EMA21"]  = frac_EMA21
Features["Frac_EMA34"]  = frac_EMA34
Features["Frac_EMA55"]  = frac_EMA55
Features["Frac_EMA89"]  = frac_EMA89
Features["Frac_EMA144"] = frac_EMA144
Features["Frac_EMA233"] = frac_EMA233
```
